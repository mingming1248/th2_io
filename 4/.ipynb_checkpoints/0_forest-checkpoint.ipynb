```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
learn_data=pd.read_csv("learn_data.csv") #读取处理好的数据
target_list=learn_data["0"].unique().tolist()#sklearn不能处理非数字，先转换一下
learn_data["0"] = learn_data["0"].apply(lambda x: target_list.index(x)) #用列表下标代替对应的字符
learn_data.groupby(["0"]).count() #查看每种算法最好时适用于多少种情况
```

```python
#使用oob，不用划分训练集和测试集
rfc=RandomForestClassifier(random_state=0,oob_score=True)
rfc=rfc.fit(learn_data[["file","per_request","sequential","requests_byte","inter_time"]],learn_data["0"])
rfc.oob_score_ 
```

```python
# D:\11-anaconda\lib\site-packages\sklearn\ensemble\_forest.py:523: UserWarning: Some inputs do not have OOB scores. This probably means too few trees were used to compute any reliable oob estimates.
# 会提示没有oobscore，暂时回避该问题，还用划分训练集和测试集的方法

# score_list = []
# for i in range(0,200,10):
#     rfc = RandomForestClassifier(n_estimators=i+1,n_jobs=-1,oob_score=True)
#     score = cross_val_score(rfc,learn_data[["file","per_request","sequential","requests_byte","inter_time"]],learn_data["0"],cv=10).mean()
#     score_list.append(score)
```

```python
X_data=learn_data[["file","per_request","sequential","requests_byte","inter_time"]]
y_data=learn_data["0"]

from sklearn.preprocessing import StandardScaler
scaler_x=StandardScaler()

X_data=scaler_x.fit_transform(X_data) #标准化，把2000缩放一下
```

```python
from sklearn.model_selection import train_test_split #划分训练集和测试集

X_train,X_test,y_train,y_test=train_test_split(X_data,y_data,test_size=0.3)

rfc=RandomForestClassifier(random_state=0)
rfc=rfc.fit(X_train,y_train)
score_r=rfc.score(X_test,y_test)

#不指定参数时的结果
print( "forest:",score_r)
```

```python
#交叉验证，看一下是否稳定

rfc=RandomForestClassifier(n_estimators=100)
rfc_s=cross_val_score(rfc,X_train,y_train,cv=10) #返回的是10个分数组成的列表

plt.plot(range(1,11),rfc_s)
plt.show()
#不太稳定，但是整体还不错
```

```python
# 随机森林调整的第一步：先来调n_estimators
estimator_list=[]
for i in range(0,200,10): #用1-200，每隔10个，不同数量的决策树组成随机森林时的效果对比，类似于学习曲线
    rfc=RandomForestClassifier(n_estimators=i+1,n_jobs=-1)
    rfc_s=cross_val_score(rfc,X_train,y_train,cv=10).mean()
    estimator_list.append(rfc_s)
```

```python
print(max(estimator_list),(estimator_list.index(max(estimator_list))*10)+1) #找到分数最大值，以及最大值时的决策树个数
plt.figure(figsize=[20,5])
plt.plot(range(1,201,10),estimator_list)
plt.show()
```

```python
#基本上在15以后就稳定了，准确率大于0.85，三次分别为61，181，191
#精细化调整n_estimators
estimator_list=[]
for i in range(50,75): #1间隔1，不同数量的决策树组成随机森林时的效果对比，类似于学习曲线
    rfc=RandomForestClassifier(n_estimators=i+1,n_jobs=-1)
    rfc_s=cross_val_score(rfc,X_train,y_train,cv=10).mean()
    estimator_list.append(rfc_s)
```

```python
print(max(estimator_list),estimator_list.index(max(estimator_list))+50) #找到分数最大值，以及最大值时的决策树个数
plt.figure(figsize=[20,5])
plt.plot(range(50,75),estimator_list)
plt.show()
```

```python
#n_estimators效果不稳定，每次结果都不一样。基本上在50以后就变化不大了，考虑计算性能，选50即可

# 为网格搜索做准备，书写网格搜索的参数
#调ax_depth
param_grid = {'max_depth':np.arange(1, 20, 1)}

rfc = RandomForestClassifier(n_estimators=50,random_state=90)

GS = GridSearchCV(rfc,param_grid,cv=10)#网格搜索
GS.fit(X_train,y_train)

print(GS.best_params_)#显示调整出来的最佳参数
print(GS.best_score_)#返回调整好的最佳参数对应的准确率
```

```python
#评分结果是在增长的，可以往泛化误差的右侧调

#调整max_features
param_grid = {'max_features':np.arange(1,5)} #默认是属性的开平方，现在往右调

rfc = RandomForestClassifier(n_estimators=50,random_state=90)

GS = GridSearchCV(rfc,param_grid,cv=10)#网格搜索
GS.fit(X_train,y_train)

print(GS.best_params_)#显示调整出来的最佳参数
print(GS.best_score_)#返回调整好的最佳参数对应的准确率
```

```python
#max_features结果反而在下降，不调也罢

#调整min_samples_leaf
param_grid={'min_samples_leaf':np.arange(1, 20, 1)}

rfc = RandomForestClassifier(n_estimators=50,random_state=90)

GS = GridSearchCV(rfc,param_grid,cv=10)#网格搜索
GS.fit(X_train,y_train)

print(GS.best_params_)#显示调整出来的最佳参数
print(GS.best_score_)#返回调整好的最佳参数对应的准确率
```

```python
#min_samples_leaf最好是6，决策树当时最好是16

#调整min_samples_split
param_grid={'min_samples_split':np.arange(2, 20, 1)}

rfc = RandomForestClassifier(n_estimators=50,random_state=90)

GS = GridSearchCV(rfc,param_grid,cv=10)#网格搜索
GS.fit(X_train,y_train)

print(GS.best_params_)#显示调整出来的最佳参数
print(GS.best_score_)#返回调整好的最佳参数对应的准确率
```

```python
#min_samples_split最好是19，决策树当时是2

#调整Criterion
param_grid = {'criterion':['gini', 'entropy']}

rfc = RandomForestClassifier(n_estimators=50,random_state=90)

GS = GridSearchCV(rfc,param_grid,cv=10)#网格搜索
GS.fit(X_train,y_train)

print(GS.best_params_)#显示调整出来的最佳参数
print(GS.best_score_)#返回调整好的最佳参数对应的准确率
```

```python
#Criterion导致的结果变差了，也抛弃

#综上所述：（第一次的数据）
#不指定参数时是0.894456289978678，但是不太稳定，整体值在0.85-0.89之间波动
#调n_estimators最好的结果是0.8715135520087134，在51时取得
#调max_depth的最好结果是0.8742595618114029，在8时取得
#调max_features的结果是0.8687696367977882，性能下降
#调min_samples_leaf的最好结果是0.8770013824305641，在6时取得
#调min_samples_split的最好结果是0.876546856017762，在19时取得
#调Criterion的最好结果是0.866936869004231，性能下降

rfc = RandomForestClassifier(n_estimators=50,random_state=90
                            ,max_depth=8
                            ,min_samples_leaf=6
                            ,min_samples_split=19
                            )
rfc=rfc.fit(X_train,y_train)
score_r=rfc.score(X_test,y_test)

#指定参数时的结果
print( "forest:",score_r)
```

```python
#综上所述：（第二次的数据）
#不指定参数时是0.8667377398720683，但是不太稳定，整体值在0.87-0.91之间波动
#调n_estimators最好的结果是0.8879665703154455，在73时取得
#调max_depth的最好结果是0.8884169075447195，在8时取得
#调max_features的结果是0.8843073185036238，在1时取得
#调min_samples_leaf的最好结果是0.8861317079301244，在3时取得
#调min_samples_split的最好结果是0.8911650119391732，在3时取得
#调Criterion的最好结果是0.8788194880817729，性能下降

rfc = RandomForestClassifier(n_estimators=50,random_state=90
                            ,max_depth=8
                            ,min_samples_leaf=3
                            ,min_samples_split=3
                            )
rfc=rfc.fit(X_train,y_train)
score_r=rfc.score(X_test,y_test)

#指定参数时的结果
print( "forest:",score_r)
```

```python
#第一次运行，指定参数时的最好结果是0.8933901918976546，不指定参数时的结果是0.894456289978678，调参后结果变差，但是差别不到
#第二次运行，指定参数时的最好结果是0.8752665245202559，不指定参数时是0.8667377398720683，调参后结果变好，且增幅明显
```

```python
for feature,importance in zip(["file","per_request","sequential","requests_byte","inter_time"],rfc.feature_importances_):
    print(feature,importance) #查看每种变量对结果的影响，没有超多30%的
    
"""
决策树的结果如下：
file 0.2578042719629814
per_request 0.4085957130709582
sequential 0.02391350771207202
requests_byte 0.03149764017683997
inter_time 0.2781888670771485
"""
```

```python
#最后，来一次全范围的网络搜索


param_grid={'max_depth':np.arange(1, 20, 1)
            ,'min_samples_leaf':np.arange(1, 20, 1)
            ,'min_samples_split':np.arange(2, 20, 1)
           }#其他参数不调整了，因为有可能让性能下降

rfc = RandomForestClassifier(n_estimators=50,random_state=90)

GS = GridSearchCV(rfc,param_grid,cv=10)#网格搜索
GS.fit(X_train,y_train)

print(GS.best_params_)#显示调整出来的最佳参数
print(GS.best_score_)#返回调整好的最佳参数对应的准确率
```

```python
rfc = RandomForestClassifier(n_estimators=50,random_state=90
                            ,max_depth=GS.best_params_["max_depth"]
                            ,min_samples_leaf=GS.best_params_["min_samples_leaf"]
                            ,min_samples_split=GS.best_params_["min_samples_split"]
                            )
rfc=rfc.fit(X_train,y_train)
score_r=rfc.score(X_test,y_test)

#用最好参数训练时的结果
print( "forest:",score_r)
```
